```python
import pandas as pd
import numpy as np
```

```python
train_set = pd.read_csv('./data/train.csv')
target = train_set['Run_time'].to_numpy()
train_set = train_set.drop(columns=['Run_time'])
```

```python
idxs = ['MWG', 'NWG', 'KWG', 'MDIMC', 'NDIMC', 'MDIMA', 'NDIMB',
         'KWI', 'VWM', 'VWN']
```

```python
for col in idxs:
    train_set[f'LOG2_{col}'] = np.log2(train_set[col])
    train_set[f'INV_{col}'] = 1.0 / train_set[col]
```
